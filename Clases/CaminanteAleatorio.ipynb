```python
def Walker(N, x0, l):

    X = []
    
    for _ in range(N):
        r = rd.random()
    #R = np.random.binomial(N,0.5)

        if r <= 0.5:
            # print("Derecha")
            x0 = x0 + l
            # print(x0)
        else:
            # print("Izquierda")
            x0 = x0 - l

        #R.append(r)
        X.append(x0)
    
    return X
```

```python
l = 1 # Paso
x0 = 0.0
N = 10000 # Número de pasos
```

```python
W = []
P = []

# Múltiples caminantes aleatorios
for i in range(100 + 1):
    w = Walker(N, x0, l) #Da únicamente un caminante aleatorio
    
    for p in w:
        P.append(p)
    plt.plot(w)
```

```python
#sns.histplot(P, bins=50, kde=True)
sns.displot(P, kind="kde")
```

```python
def Walker2(N, x0, l):

    X = []
    
    for _ in range(N):
        r = np.random.normal(0.0, np.pi)
    #R = np.random.binomial(N,0.5)

        if r <= 0.0:
            # print("Derecha")
            x0 = x0 + l
            # print(x0)
        else:
            # print("Izquierda")
            x0 = x0 - l

        #R.append(r)
        X.append(x0)
    
    return X
```

```python
l = 1 # Paso
x0 = 0.0
N = 10000 # Número de pasos

W = []
P = []

for i in range(100 + 1):
    w = Walker2(N, x0, l)
    
    for p in w:
        P.append(p)
    plt.plot(w)
```
