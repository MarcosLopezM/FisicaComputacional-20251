```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
})
```

```python
def dev_delante(f, x, h):
    return (f(x + h) - f(x)) / h

def dev_atras(f, x, h):
    return (f(x) - f(x - h)) / h

def dev_central(f, x, h):
    return (f(x + h) - f(x - h)) / (2*h)
```

```python
f1 = lambda x: x**2 + 3 * x + 8
f2 = lambda x: 3*np.exp(x) + (np.sin(x))**2
```

```python
h = 1e-12
x = np.linspace(-10, 10, 1000)

fig, axs = plt.subplots(3, 1, figsize=(8, 12), sharex=True)

titles = ["Diferencia hacia adelante", "Diferencia hacia atrás", "Diferencia Central"]
derivatives = [dev_delante, dev_atras, dev_central]
colors = ['g', 'magenta', 'crimson']

for i, (ax, title, deriv_func, color) in enumerate(zip(axs, titles, derivatives, colors)):
    ax.plot(x, f1(x), color="indigo", label="$f(x) = x^{2} + 3x + 8$")
    ax.plot(x, deriv_func(f1, x, h), color=color, label="$f^{\\prime}(x)$")
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    ax.set_title(title)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
h = 1e-12
x = np.linspace(-10, 10, 1000)

fig, axs = plt.subplots(3, 1, figsize=(8, 12), sharex=True)

titles = ["Diferencia hacia adelante", "Diferencia hacia atrás", "Diferencia Central"]
derivatives = [dev_delante, dev_atras, dev_central]
colors = ['g', 'magenta', 'olive']

for i, (ax, title, deriv_func, color) in enumerate(zip(axs, titles, derivatives, colors)):
    ax.plot(x, f2(x), label="$f(x) = 3e^{x} + \\sin^{2}(x)$", color="indigo")
    ax.plot(x, deriv_func(f2, x, h), '--', color=color, label="$f^{\\prime}(x)$", ms=5)
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    ax.set_title(title)
    ax.legend()

plt.tight_layout()
plt.show()
```
