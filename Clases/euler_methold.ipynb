<h1>MÃ©todo de Euler</h1>

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
})
```

```python
def euler_method(f, y_init, x):
    y_i = y_init
    y = [y_i]
    for i in range(len(x) - 1):
        y.append(f(y[i]) * (x[i + 1] - x[i]) + y[i])
    return y
```

```python
f = lambda y: y
x = np.linspace(0, 4, 1000)
y_0 = 1

y = euler_method(f, y_0, x)
```

```python
plt.figure(figsize=(12, 7))
plt.plot(x, y)
plt.title("SoluciÃ³n de $y^{\\prime} = y$ por el mÃ©todo de Euler")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.grid(alpha=0.3)
plt.show()
```
