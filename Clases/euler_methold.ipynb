<h1>Método de Euler</h1>

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
})
```

```python
def euler_method(f, y_init, x):
    y_i = y_init
    y = [y_i]
    for i in range(len(x) - 1):
        y.append(y[i] + (x[i + 1] - x[i]) * f(y[i]))
    return y
```

```python
f = lambda y: y
x = np.linspace(0, 4, 1000)
y_0 = 1

y = euler_method(f, y_0, x)
```

```python
plt.figure(figsize=(12, 7))
plt.plot(x, y)
plt.title("Solución de $y^{\\prime} = y$ por el método de Euler")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.grid(alpha=0.3)
plt.show()
```

```python
def Euler2D(f, g, x_init, y_init, t_init, t_end, delta_t):
    X = []
    Y = []
    T = []

    X.append(x_init)
    Y.append(y_init)
    T.append(t_init)

    x = x_init
    y = y_init
    t = t_init

    while t < t_end:
        m = f(x, y)
        n = g(x, y)

        x += m * delta_t
        y += n * delta_t

        t += delta_t

        X.append(x)
        Y.append(y)
        T.append(t)

    return X, Y, T
```
