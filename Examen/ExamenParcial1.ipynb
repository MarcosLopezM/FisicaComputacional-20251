<h2>Problema 2</h2>

```python
def factorial(n):
    if n < 2:
        return 1
    return n * factorial(n - 1)
```

```python
def aprox_pi(k):
    approx = 0
    for i in range(k):
        approx += ((2**i)*(factorial(i)**2)) / factorial(2*i + 1)
    return approx
```

```python
aprox_pi(1000)
```

<h2>Problema 3</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import loadtxt

# Datos de un movimiento rectilíneo uniforme en un intervalo de timepo [t_i, t_f]
archivo = open("mru.txt", "w")

def mov_rect(x_init, v_init, t):
    return x_init + v_init*t

# Condiciones iniciales
x_init = 0
v_init = 6.66
t_init = 0.
t_final = 6.
t_values = np.linspace(t_init, t_final, 1000)

# Escritura de los datos
for t in t_values:
    archivo.write(f"{mov_rect(x_init, v_init, t)}\n")

# Lectura de archivos y graficación
mru_data = loadtxt("mru.txt")

plt.figure(figsize=(8,8))
plt.plot(t_values, mru_data, 'b')
plt.grid()
plt.title("Movimiento rectilíneo uniforme")
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```
